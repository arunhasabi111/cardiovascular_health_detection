```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
Heart_disease = pd.read_csv("framingham_heart_disease.csv")
```

```python
Heart_disease.head()
```

```python
Heart_disease.info()
```

```python
Heart_disease.columns
```

```python
Heart_disease.isnull().sum()
```

```python
HD = Heart_disease.drop(["education"],axis=1)
HD.head()
```

```python
mean_cigsPerDay = round(HD["cigsPerDay"].mean())
mean_BPMeds = round(HD["BPMeds"].mean())
mean_totChol = round(HD["totChol"].mean())
mean_heartRate = round(HD["heartRate"].mean())
mean_BMI = round(HD["BMI"].mean())
mean_glucose = round(HD["glucose"].mean())
```

```python
HD["cigsPerDay"].fillna(mean_cigsPerDay, inplace= True)
HD["BPMeds"].fillna(mean_BPMeds, inplace= True)
HD["heartRate"].fillna(mean_heartRate, inplace= True)
HD["totChol"].fillna(mean_totChol, inplace= True)
HD["BMI"].fillna(mean_BMI, inplace= True)
HD["glucose"].fillna(mean_glucose, inplace= True)
```

```python
HD.isnull().sum()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(HD.corr(), annot=True, cmap='terrain')
```

```python
X = HD.drop('TenYearCHD',axis=1)
Y = HD["TenYearCHD"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size= 0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()

classifier.fit(X_train,Y_train)
```

```python
Y_test_HD = classifier.predict(X_test)
```

```python
result = pd.DataFrame({'Actual': Y_test, 'Prediction': Y_test_HD})
```

```python
result.head(5)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_test_HD)
```

```python
Y_train_HD = classifier.predict(X_train)
Y_train_HD
```

```python
accuracy_score(Y_train, Y_train_HD)
```

```python
Y_test_HD_proba = classifier.predict_proba(X_test)
Y_test_HD_proba
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, Y_test_HD)
```

```python
cm
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot= True)

plt.title("Confusion Matrix - Test Data")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
```

```python
plt.imshow(cm, interpolation = 'nearest', cmap = plt.cm.Wistia)

classNames = ['Heart Disease', 'NO Heart Disease']
plt.title('Confusion Matrix-Test Data')
plt.ylabel('True label')
plt.xlabel('Predicted label')

tick_marks = np.arange(2)
plt.xticks(tick_marks, classNames, rotation =45)
plt.yticks(tick_marks, classNames)

s=[['TN','FP'],['FN','TP']]
for i in range(2):
    for j in range(2):
        plt.text(j, i, str(s[i][j])+"= "+str(cm[i][j]))
        
        
plt.show();
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_test_HD))
```

```python
TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TP = cm[1][1]
```

```python
recall = TP/(TP+FN)
print("Recall= ",recall)
```

```python
Precision = TP/(TP+FP)
print("Precision= ",Precision)
```

```python
specificity = TN/(TN+FP)
print("Specificity= ",specificity)
```

```python
accuracy = (TP+TN)/(TN+TP+FP+FN)
print("Accuracy= ",accuracy)
```
